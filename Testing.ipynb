# video_summarization

```python
def frames_extraction(video_path):
    IMAGE_HEIGHT , IMAGE_WIDTH = 224, 224
    count=0
    video_reader = cv2.VideoCapture(video_path)
    video_frames_count = int(video_reader.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = video_reader.get(cv2.CAP_PROP_FPS)
    print(video_frames_count)
    SEQUENCE_LENGTH = video_frames_count
    print("Frames per second using video.get(cv2.cv.CV_CAP_PROP_FPS): {0}".format(fps))
    skip_frames_window = max(int(video_frames_count/SEQUENCE_LENGTH), 5)
    success, frame1 = video_reader.read() 
    prev_frame = frame1
    threshold=20.
    
    a = 0
    b = 0
    c = 0

    while True:
        success, frame = video_reader.read() 
        if success is True:
            if (np.sum((np.absolute(frame-prev_frame))/np.size(frame)) > threshold):
                resized_frame = cv2.resize(frame, (IMAGE_HEIGHT, IMAGE_WIDTH))
                prev_frame = frame
                a += 1
            else:
                prev_frame = frame
                b += 1
            c += 1
            frame=resized_frame.reshape(-1,224,224,3)
            prob=np.argmax(model.predict(frame))
            
        if prob == 0:
            print("Anglectesia")
            cv2.imwrite("./saved_frames/Anglectesia%d.png" % count, np.squeeze(frame,0))
        elif prob == 1:
            print("esophagitis")
            cv2.imwrite("./saved_frames/esophagitis%d.png" % count, np.squeeze(frame,0))
        elif prob == 2:
            print("Foreign body")
            cv2.imwrite("./saved_frames/Foreign_body%d.png" % count, np.squeeze(frame,0))
        elif prob == 3:
            print("Normal")
            pass
            #cv2.imwrite("Normal%d.png" % count, frame)

        else:
            print("Ulcer")
            cv2.imwrite("./saved_frames/Ulcer%d.png" % count, np.squeeze(frame,0))
        count+=1
        
        
    video_reader.release()
```
